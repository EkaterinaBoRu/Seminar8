## 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df_kc.hist(figsize=(16, 16), bins=20, grid=False);
```

```python
corr_matrix = df_kc.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

Вывод: из тепловой карты видно, что на цену недвижимости оказывают влияние: количество спален, ванных комнат, площадь дома, количество этажей, наличие вида на набережную и внешний вид дома, класс жилья, площадь наземной и подземной части дома, широта, средняя площадь прилегающих домов

```python
sns.jointplot(x=df_kc['sqft_living'], y=df_kc['price'], kind='reg');
```

Вывод: наблюдается линейная зависимость - с увеличением площади дома повышается стоимость данного дома

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df_kc['grade'], y=df_kc['price'], whis=0.3)

plt.xlabel('Класс жилья')
plt.ylabel('Стоимость')
plt.title('Зависимость стоимости дома от класса жилья');
```

Вывод: чем выше класс жилья - тем выше цена недвижимости. С увеличением класса жилья увеличивается и диапазон цен на недвижимость внутри класса.

```python
plt.figure(figsize=(6, 4))

plt.scatter(df_kc['sqft_above'], df_kc['price'])

plt.title('Зависимость стоимости дома от площади надземной части дома')
plt.xlabel('Площадь')
plt.ylabel('Стоимость');
```

Вывод: стоимость дома прямо пропорциональна площади наземной части дома

```python
plt.figure(figsize=(6, 4))

sns.lineplot(x=df_kc['bathrooms'], y=df_kc['price'])

plt.title('Зависимость стоимости дома от количества ванных комнат')
plt.xlabel('Ванные комнаты')
plt.ylabel('Стоимость дома')
```

Вывод: наблюдается прямая линейная зависимость, что стоимость недвижимости увеличивается с увеличением количества санитарных узлов в доме
